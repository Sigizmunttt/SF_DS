```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
parsing=pd.read_excel("Data_TSUM.xlsx",sheet_name=[0])
company=pd.read_excel("Data_TSUM.xlsx",sheet_name=[1])
z=["_","-","~","\\","/"]

parcing=pd.read_excel("Data_TSUM.xlsx",sheet_name="Data_Parsing")
company=pd.read_excel("Data_TSUM.xlsx",sheet_name="Data_Company")
parcing = parcing.replace(to_replace= ['_', '-', '~', '/', '\\\\'] , value='', regex=True)
parcing=parcing.rename(columns={"producer_id":"item_id","producer_color":"color_id"})

merged = parcing.merge(
    company,
    on=['item_id',"color_id"],
    how='left'
)

parcing["difference%"]=round(merged["price"]/merged["current price"]*100 - 100,2)
fig = plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(
    data=parcing,
    y='brand',
    x='difference%',
    orient='h',
    width=0.9
)
boxplot.set_title('Распределение летальности по странам');
boxplot.set_xlabel('Летальность');
boxplot.set_ylabel('Страна');
boxplot.grid()
```
